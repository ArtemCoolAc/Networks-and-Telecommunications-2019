```python
import random
import numpy
import time
import this
```

```python
def generate(p): # генератор с заданной вероятностью
    x = random.random()
    if x<=p:
        return True
    else:
        return False
    return False
```

```python
s=0
quantity = 100000000
for i in range(quantity):
    t = generate(0.7)
    s+=t
    #print(t)
print("Итоговая вероятность: {}".format(s/quantity))
```

```python
class Situation:
    def __init__(self, packets_quantity, packets_limit, lose_probability):
        self.start_packets_quantity = packets_quantity # изначальное количество пакетов, на которое делится сообщение
        self.maximum_packets_limit = packets_limit # общий стартовый лимит на количество передач пакетов
        self.current_packets_quantity_to_confirm = packets_quantity # оставшееся количество пакетов для передачи
        self.current_packets_limit = packets_limit # оставшееся количество количество попыток для передачи
        self.success_probability = 1 - lose_probability # вероятность успешной доставки
        self.lose_probability = lose_probability # вероятность неуспешной доставки
        self.confirmed_packets = list() # список уже подтвержденных пакетов
        self.packets_to_confirm = list() # список ещё не подтвержденных пакетов
        self.full_success = 0
        self.failed = False
        for i in range(self.start_packets_quantity):
            self.packets_to_confirm.append(i)
        print(self.start_packets_quantity)
        print(self.maximum_packets_limit)
        print(self.current_packets_quantity_to_confirm)
        print(self.current_packets_limit)
        print(self.success_probability)
        print(self.lose_probability)
        print(self.confirmed_packets)
        print(self.packets_to_confirm)
        print(self.full_success)
        
    def print_attributes(self):
        print(self.start_packets_quantity)
        print(self.maximum_packets_limit)
        print(self.current_packets_quantity_to_confirm)
        print(self.current_packets_limit)
        print(self.success_probability)
        print(self.lose_probability)
        print(self.confirmed_packets)
        print(self.packets_to_confirm)
        print(self.full_success)
    
    def sending_a_message(self):
        unconfirmed_packets_quantity = len(self.packets_to_confirm)
        print("Подтвержденные пакеты до попытки: {}".format(self.confirmed_packets))
        print("Неподтвержденные пакеты до попытки отправки: {}".format(self.packets_to_confirm))
        preliminary_list_of_cuccess_packets = list()
        for i in range(unconfirmed_packets_quantity):
            this_cycle_confirmed_packets = 0
            packet_is_delivered = generate(self.success_probability)
            print(packet_is_delivered)
            print("Пакет {} был {}".format(self.packets_to_confirm[i], packet_is_delivered))
            self.current_packets_limit -= 1
            if packet_is_delivered:
                this_cycle_confirmed_packets += 1
                preliminary_list_of_cuccess_packets.append(self.packets_to_confirm[i])
                print("Это был пакет {}".format(self.packets_to_confirm[i]))
        print("Предварительный список после цикла: {}".format(preliminary_list_of_cuccess_packets))
        special_packet_delivered = generate(self.success_probability)
        print("Сообщение о подтверждении было {}".format(special_packet_delivered))
        if special_packet_delivered:
            self.confirmed_packets = self.confirmed_packets + preliminary_list_of_cuccess_packets
            print("Подтвержденные пакеты: {}".format(self.confirmed_packets))
            self.packets_to_confirm = list(set(self.packets_to_confirm) - set(preliminary_list_of_cuccess_packets))
            ###print("Ещё не подтвержденные пакеты")
            self.current_packets_quantity_to_confirm = len(self.packets_to_confirm)
            print("Ещё не подтвержденные пакеты: {}".format(self.current_packets_quantity_to_confirm))
            if self.current_packets_quantity_to_confirm == 0 and self.current_packets_limit >= 0:
                self.full_success = 1
            if self.current_packets_limit < 0:
                self.failed = True
                print("ЭТО EPIC FAIL!")
        return self.failed
    
    def full_messaging(self):
        print("Начинаем отправку сообщения")
        while self.current_packets_limit >= 0 and self.failed == False and self.full_success != 1:
            fail = self.sending_a_message()
            if fail:
                self.full_success = 0
            print("Утверждение, что сообщение доставлено успешно {}, осталось попыток чтобы закончить {}".format(self.full_success, self.current_packets_limit))
        return self.full_success
```

```python
cur_sum = 0
for i in range(100):
    model = Situation(5,18,0.615712416801)
    s1 = model.full_messaging()
    cur_sum += s1
    print("Смогли закончить {}".format(s1))
print("Итогово удачных попыток {}".format(cur_sum))
```
