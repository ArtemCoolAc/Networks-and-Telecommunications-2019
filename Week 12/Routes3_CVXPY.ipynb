```python
paths = list(nx.all_simple_paths(G,1,10))
```

```python
pprint.pprint(paths)
```

```python
matrix = [[0 for j in range(len(edges))] for i in range(len(paths))]
pprint.pprint(matrix)
for i in range(len(paths)):
    for j in range(len(paths[i])-1):
        current = paths[i][j]
        next1 = paths[i][j+1]
        index = edges.index([current, next1])
        matrix[i][index] = 1
pprint.pprint(matrix)
```

```python
eff_matrix = [[0 for j in range(len(edges))] for i in range(len(paths))]
#pprint.pprint(eff_matrix)
for i in range(len(paths)):
    for j in range(len(paths[i])-1):
        current = paths[i][j]
        next1 = paths[i][j+1]
        index = edges.index([current, next1])
        eff_matrix[i][index] = 1 * a[index]
pprint.pprint(eff_matrix)
```

```python
global_string = ''
global_string += '['
for i in range(len(eff_matrix)):
    string = ''
    string += '['
    for j in range(len(eff_matrix[i])):
        if eff_matrix[i][j] != 0:
            string += '+a[{}]*r[{}]'.format(j,j)
    string += '],\n'
    string = '[' + string[2:]
    #print(string)
    global_string += string
print(global_string)
print()
global_string = global_string[:-1] + ']'
print(global_string)
```
