```python
def find_all_paths(graph, start, end, path=[]):
    path = path + [start]
    if start == end:
        return [path]
    if start not in graph:
        return []
    paths = []
    for node in graph[start]:
        if node not in path:
            newpaths = find_all_paths(graph, node, end, path)
            for newpath in newpaths:
                paths.append(newpath)
    return paths
```

```python
paths = find_all_paths(g,1,50)
pprint.pprint(paths)
```

```python
def count_one_probability(number, effectivity):
    summ = 0
    for i in range(len(paths[number])-1):
        current = paths[number][i]
        next1 = paths[number][i + 1]
        #print('Путь {} -> {})'.format(current, next1))
        summ += effectivity[current][next1]
    #print(summ)
    return summ
```

```python
effect_paths = []
for i in range(len(paths)):
    eff = count_one_probability(i, eff_dict)
    effect_paths.append(eff)
print(effect_paths)
```
