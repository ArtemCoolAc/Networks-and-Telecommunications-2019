```python
def optPath(G,p):
    edge2ind = {e:i for i,e in enumerate(G.edges())}
    H = G.copy()
    p = np.minimum(p,1)
    w = -np.log(p)
    n = H.number_of_nodes()
    
    for i in H:
        for j in H[i]:
            H[i][j]['w'] = w[edge2ind[(i,j)]]
            
    path = nx.shortest_path(H,1,n,'w')
    foo = [edge2ind[(i,j)] for i,j in zip(path[:-1],path[1:])]
    x = np.zeros_like(p)
    x[foo] = 1
    return x
```
